#### finding best result for dog breed in data of the neural network

If the neural network found a dog breed at the first hit, it is inserted into the data set under the variable "dog". Otherwise, the system checks whether the neural network was able to land a hit at position 2 or 3. For any questions that may arise during the analysis, the probability with which the identified dog breed can actually be seen on the photo is also included in the data set. 

```python
# finding the best dog-related solution of the neuronal network 
def dogtest(x): 
    if x["p1_dog"] == True: 
        return pd.Series ([x["p1"], x["p1_conf"]])
    elif x["p2_dog"] == True: 
        return pd.Series ([x["p2"], x["p2_conf"]])
    elif x["p3_dog"] == True:
        return pd.Series ([x["p3"], x["p3_conf"]])
    else:
        return pd.Series ([None, None])

dog_copy[["dog", "dog_prob"]] = dog_copy.apply(dogtest, axis=1)
```

```python
#  get rid of the unnecessasry columns in dog_copy
dog_copy = dog_copy[["tweet_id", "dog", "dog_prob"]]
```

#### Making the data of dog_types tidy

The data of the respective dog type (doggo, floofer, pupper, puppo) in the Twitter archive are not tidy. Instead of having a separate column for each dog type, we create the variable dog_type, under which the respective dog type is listed. If no dog type is specified, a "None" is entered. 

```python
lib_copy.shape
```

```python
lib_copy.head(1)
```

```python
# first we have to check for every row, if any of the dog-type-variables is filled out.
# If not, we are setting the variable dog_type to "None"

def check_dog_type(x): 
    
    if x["doggo"] != "None":
        return "doggo"
    elif x["floofer"] != "None":
        return "floofer"
    elif x["pupper"] != "None": 
        return "pupper"
    elif x["puppo"] != "None": 
        return "puppo"
    else: 
        return "None"

lib_copy["dog_type"] = lib_copy.apply(check_dog_type, axis=1)

    
```

```python
# delete the unnecessary rows doggo, floofer, pupper, puppo
lib_copy.drop(["doggo", "floofer", "pupper", "puppo"], axis=1, inplace=True)
```

#### What is the average rating for the ten most popular dog breeds (according to tweets)? 

```python
# Finding the average rating for the 10 most popular dog breeds
df_populardogs.agg({"tweet_id":"size", "rating_numerator":"mean"})\
.rename(columns={"tweet_id":"count", "rating_numerator":"rating numerator"})\
.sort_values(by="count", ascending=False)\
.head(10)
```

#### What are the five most popular dog names? What count of retweets and wha erating numerator do they have in mean?

```python
df_popularnames = tam[tam["name"] != "None"].groupby("name")
df_popularnames.count().sort_values(by="tweet_id", ascending=False)

df_popularnames.agg({"tweet_id":"size", "rating_numerator":"mean", "retweet_count":"mean"})\
.rename(columns={"tweet_id":"count", "rating_numerator":"rating numerator", "retweet_count":"retweets"})\
.sort_values(by="count", ascending=False)\
.head(5)
```

#### 7. Which ten dog breeds with at least five entries in our data set are rated worst? 

```python
not_all_good_dogs = tam[["dog","tweet_id", "rating_numerator", "retweet_count"]]
not_all_good_dogs = not_all_good_dogs.groupby("dog")\
.agg({"rating_numerator":"mean", "tweet_id":"size", "retweet_count":"mean"})\
.rename(columns= {"tweet_id":"count", "rating_numerator": "rating", "retweet_count": "retweets"})
not_all_good_dogs[not_all_good_dogs["count"]>=5].sort_values(by="rating").head(10)
```
